```python
# File mergedexc.py (Python 3.X + 2.X)
sep = '-' * 45 + '\n'
```

En el código se intenta acceder al índice 99 de "spam" lo que genera un error "IndexError". La siguiente linea indica que si existe un "IndexError" nos tiene que devolver "except run" el cual en este código pasa. Luego hay un "finally" el cual siempre se ejecuta pase lo que pase imprimiendo "finally run" y por último estaría print que nos imprime un "after run"

```python
print(sep + 'EXCEPTION RAISED AND CAUGHT')
try:
    x = 'spam'[99]
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')
```

En el código se intenta acceder al índice 3 de "spam" no habiendo problema. La siguiente linea indica que si existe un "IndexError" nos tiene que devolver "except run" sin embargo el la linea anterior no dió ningun fallo por lo que esta accion queda invalidada. Despues se añade debajo un else el cual funciona si no salta la excepción de "IndexError" haciendo que nos imprima un "else run". Luego hay un "finally" el cual siempre se ejecuta pase lo que pase "finally run" y por último estaría print que nos imprime un "after run"

```python
print(sep + 'NO EXCEPTION RAISED, WITH ELSE')
try:
    x = 'spam'[3]
except IndexError:
    print('except run')
else:
    print('else run')
finally:
    print('finally run')
print('after run')
```

En el código empieza con "x = 1 / 0" el cual genera un error. La siguiente linea indica que si existe un "IndexError" nos tiene que devolver "except run" sin embargo el la linea anterior tiene un error distinto al que hemos puesto en "except". Despues se añade debajo un else el cual funciona si no salta la excepción de "IndexError" haciendo que nos imprima un "else run". Luego hay un "finally" el cual siempre se ejecuta pase lo que pase "finally run" y por último estaría print que nos imprime un "after run"

```python
print(sep + 'EXCEPTION RAISED BUT NOT CAUGHT')
try:
    x = 1 / 0
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')
```
