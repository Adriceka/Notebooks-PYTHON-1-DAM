## OBJETOS MUTABLES

### Listas

En este codigo hay dos valores, una que es "L" el cual tiene una lista mutable de 3 numeros (2,3 y 4) y luego tenemos el valor "M" que le decimos que tenga el valor de "L".
Luego de asignar los valores actualizamos la "L" poniendo "L[0] = 24" diciendo que de la lista anterior remplacemos el numero que este en la posición 0 empezando desde la izquierda por el numero 24. Una vez hecho esto le pedimos que nos de el valor de "L" y "M" dandonos el valor de la lista actualizada de "L" siendo ahora [24,3,4] en vez de [2,3,4]. Sin embargo "M" tiene la lista actualizada de "L" porque a diferencia de los codigos anteriores donde habria que pones de nuevo que "L" es igual a "M" estas comparten la misma memoria apuntando ambas a la misma lista haciendo que si tu actualizas la lista de "M" tambien se actualice la lista de "L"

```python

L = [2, 3, 4]	    # A mutable object
M = L			    # Make a reference to the same object
L[0] = 24 		    # An in-place change

print("L = ", L)    # [24, 3, 4]
print("M = ", M)    # [24, 3, 4]
```

Copia de objetos en vez de referencia:


Este código es igual al anterior añadiendo una pequeña cosa. En el ejercicio anteiror estaba puesto que "M = L" pero ahora en este codigo añadimos "M = L[:]". En el resultado anterior nos daba la misma lista porque ambos valores apuntaban a la misma lista dando igual si actualizabas "L" o "M" pero este caso "L" se queda con la lista actualizada y "M" tiene la lista antigua que tenia "L". Esto se debe que al poner M = L[:] en vez que ambos valores apunten a la misma lista le decimos a "M" que haga una copia de la lista que tenga "L" para si misma apuntando a otra dirección distinta. Por eso cuando actualizamos el valor de "L" la "M" se queda con la misma lista porque tiene la suya propia

```python

L = [2, 3, 4]
M = L[:]		    # Make a copy of L (or list(L), copy.copy(L), etc.)
L[0] = 24

print("L = ", L)    # [24, 3, 4]

print("M = ", M)    # [2, 3, 4]  M is not changed
```

### Shared References and Equality
- Operador `==` igualdad de valores.
- Operador `is` identidad de objetos.

En este código vuelven a estar los mismos valores "M" y "L" teniendo "L" la lista y luego en el codigo decimos que "M = L" para que ambas apunten a la misma lista. Depsues ponemos en el codigo "L == M" y "L is M" y nos devuelve true. Esto es porque le preguntamos al código con el "==" si "M" y "L" comparten los mismos valores siendo verdadero porque comparten la misma lista. Luego al poner "is" preguntamos si "L" y "M" estan apuntando a la misma lista de objetos siendo tambien verdadero. Por eso dos devuelve un "true" para ambas preguntas

```python

L = [1, 2, 3]
M = L 			    # M and L reference the same object

L == M 			    # True: Same values => operador igualdad de valores
L is M 			    # True: operador identidad de objetos: compara las referencias (los punteros)
```

Este códigop es parecido al anterior pero ahora en vez de tener "L" y "M" en la misma lista cada una tiene la suya propia pero con los mismo valores. Ahora al preguntarle con "==" e "is" nos devuelve un verdadero y falso. Esto se debe porque al poner "L == M" le preguntamos si los dos valores comparten los mismos valores siando True en este caso pero en "L is M" le estamos a preguntar si ambos valores apuntan a la misma lista de objetos siendo False porque cada valor tiene una lista propia apuntndo a distintas direcciones. Por eso nos dá primero un True porque comparten los mismos valores pero un False porque no apuntan a la misma lista.

```python
L = [1, 2, 3]
M = [1, 2, 3]	    # M and L reference different objects

print(
    "L == M",
    L == M 			# True: Same values
)

print(
    "L is M", 	    # False: operador identidad de objetos: compara las referencias (los punteros)
    L is M
)
```

#### CACHE

Because small integers and strings are cached and reused, though, is tells us they reference the same single object.


En este codigo asignamos a dos valores "X" e "Y" el mismo valor siendo 42 y le hacemos la pregunta como los ejercicios anteirores de la lista con el "==" y el "is". En este caso al preguntar "X == Y" nos dice que es True porque ambos tienen el mismo valor y al preguntar "X is Y" nos da true aunque le asignemos el valor por separado a cada uno. Esto a diferencia de las listas al ser codigo pequeño con poca información se almacena ambos valores en la caché y como los dos comparten el mismo número apuntan a la misma dirección

```python

X = 42
Y = 42		    # Should be two different objects

X  == Y         # True
X is Y 		    # True: Same object anyhow: caching at work!!
```
